```
def plot_random_spectrogram(self,name,save = False,limits =(0,20000) ):
    plt.figure(figsize=(14, 5))
    X = librosa.stft(self.signalData)
    Xdb = librosa.amplitude_to_db(abs(X))
    plt.figure(figsize=(14, 5))
    plt.title(str(name))
    librosa.display.specshow(Xdb, sr=self.samplingFrequency, x_axis='time', y_axis='hz')
    plt.colorbar(format='%+2.0f dB')
    if save == True:
        plt.savefig('spectogram-'+str(name)+'.png')
```

```
def plot_random_spectrogram_limited(self,name,save = False,limits =(0,20000) ):
    plt.figure(figsize=(14, 5))
    X = librosa.stft(self.signalData)
    Xdb = librosa.amplitude_to_db(abs(X))
    plt.figure(figsize=(14, 5))
    plt.title(str(name))
    librosa.display.specshow(Xdb, sr=self.samplingFrequency, x_axis='time', y_axis='hz')
    plt.colorbar(format='%+2.0f dB')
    l1,l2 = limits
    plt.ylim(l1,l2)
    if save == True:
        plt.savefig('spectogram-limited-'+str(name)+'.png')
for no,i in enumerate(wavfiles):
    f = audio(i)
    plot_random_spectrogram_limited(f,no,save = True,limits =(0,2000) )
```
